## <font color=darkblue> Plotting a Binomial tree </font> 

**Binomial tree for n=6**

```python
from math import *
import math as m;
import pylab as z;
import numpy as np;
import pandas as pd;
import scipy.stats  as stats;
```

Defining the volatility of the the binomial tree <br>
u = 1.30 <br>
d = 1/u 

**Defining the functions to calculate the prices**

```python
def binomial(d, u, p):
    g = np.random.binomial(1,p)

    if g == 1:
        return u
    else:
        return d
```

**User input for the variables**

```python
nodes = 6 #Nodes
S = 100.0 #Initial spot price
u = 1.30 #Up factor
d = 1/u #Down factor
n = 5 #Steps
p = 0.5 #Probability of up and down moves 50% each


numberofpaths = 2**nodes
valuelist = [] #Emptylist
z.figure(0) #Generating the figure
temp = S #Temporary variable
```

**Calculating and plotting**

```python
for i in range(0, numberofpaths, 1):
    valuelist =[]
    S = temp;
    for c in range(0, nodes + 1, 1):
        valuelist.append(S)
        S = S*binomial(d, u, p)
    
    z.plot(range(0, nodes + 1, 1), valuelist)
    for i in range(nodes):
        z.text(i+.2,valuelist[i]-1,'{:4.1f}'.format(valuelist[i]))


z.title("Price movements")
z.xlabel("Nodes")
z.ylabel("S(n)")
z.show()
```
